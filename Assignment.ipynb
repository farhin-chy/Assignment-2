```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
X = np.array([1, 2, 3, 4, 5], dtype=float)  # size
y = np.array([150, 200, 250, 300, 350], dtype=float)  # price

# Initialize parameters
m = 0.0  # slope
b = 0.0  # intercept
lr = 0.01
epochs = 200

losses = []
accuracies = []

# Gradient Descent
for epoch in range(epochs):
    # Prediction
    y_pred = m * X + b

    # Error
    error = y_pred - y
    mse = np.mean(error**2)
    losses.append(mse)

    # Define accuracy (pseudo accuracy based on % error)
    accuracy = 100 - (np.mean(np.abs(error / y)) * 100)
    accuracies.append(accuracy)

    # Gradients
    dm = np.mean(2 * error * X)
    db = np.mean(2 * error)

    # Update parameters
    m -= lr * dm
    b -= lr * db

    if epoch % 20 == 0:
        print(f"Epoch {epoch}, Loss={mse:.4f}, Accuracy={accuracy:.2f}%")

print(f"\nâœ… Final Model: price = {m:.2f}*size + {b:.2f}")

# ------------------ Plot 1: Epoch vs Accuracy ------------------
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(epochs), accuracies, label="Accuracy", color="green")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.title("Epoch vs Accuracy for Linear Regression")
plt.legend()

# ------------------ Plot 2: Data vs Regression Line ------------------
plt.subplot(1, 2, 2)
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, m * X + b, color="red", label="Regression Line")
plt.xlabel("House Size (X)")
plt.ylabel("Price (y)")
plt.title("House Size vs Price with Regression Line")
plt.legend()

plt.tight_layout()
plt.show()
```
